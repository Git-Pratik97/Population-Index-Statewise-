```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import matplotlib.gridspec as gridspec
import matplotlib as mpl
mpl.get_backend()
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
```

```python
df = pd.read_csv('population_india_census2011.csv')
df.to_csv('population_india_census2011.csv', index=False)
data = pd.DataFrame(df)
data.index

data.set_index("Sno")
data.head()
```

```python
#Population Comparison according the data given...
fig = plt.figure()
x1 = plt.plot(data["State / Union Territory"],data["Population"]*10**(-8),"--r",data["State / Union Territory"], data["Rural population"]*10**(-8),"-o",data["State / Union Territory"], data["Urban population"]*10**(-8), "-x")
plt.legend(["Popalation","Rural Population", "Urban Population"])
ax = plt.gca()
ax = [0,20,0,0.5]
plt.xlabel("State / Union Territory")
plt.ylabel("Population")
plt.title("Statewise Population")
x = plt.gca().xaxis
for i in x.get_ticklabels():
    i.set_rotation(90)
plt.subplots_adjust(bottom=-2, left= -2)
```

```python
##Region Between the Rural and Urban Population;
plt.gca().fill_between(range(len(data)), 
                       data["Rural population"], data["Urban population"], 
                       facecolor='blue', 
                       alpha=0.25);
```
